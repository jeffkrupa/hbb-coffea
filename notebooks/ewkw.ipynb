```python
from coffea.nanoevents import NanoEventsFactory
import awkward as ak
import numpy as np
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings("ignore", "Found duplicate branch")
```

```python
had_file = "root://cmsxrootd.fnal.gov//store/group/lpcpfnano/jdickins/v2_2/2016/EWKV/EWKWminus2Jets_WToQQ_dipoleRecoilOn_TuneCP5_13TeV-madgraph-pythia8/EWKWminus_WToQQ/220530_134431/0000/nano_mc2016post_1-1.root"

had_events = NanoEventsFactory.from_root(
    had_file,
    metadata={"dataset": "EWKWMinus2Jets_WToQQ"},
    #entry_stop=100000,
).events()
```

```python
lep_file = "root://cmsxrootd.fnal.gov//store/group/lpcpfnano/jdickins/v2_2/2016/EWKV/EWKWMinus2Jets_WToLNu_M-50_TuneCP5_withDipoleRecoil_13TeV-madgraph-pythia8/EWKWminus_WToLNu/220315_212930/0000/nano_mc2016post_1.root"

lep_events = NanoEventsFactory.from_root(
    lep_file,
    metadata={"dataset": "EWKWMinus2Jets_WToLNu"},
    #entry_stop=100000,
).events()
```

```python
had_bquarks = ak.firsts(had_events.GenPart.pt[abs(had_events.GenPart.pdgId) == 5])
lep_bquarks = ak.firsts(lep_events.GenPart.pt[abs(lep_events.GenPart.pdgId) == 5])
```

```python
plt.hist(had_bquarks[had_bquarks>0],histtype='step',bins=np.linspace(10,500,50),label='WToQQ',density=True)
plt.hist(lep_bquarks[lep_bquarks>0],histtype='step',bins=np.linspace(10,500,50),label='WToLNu',density=True)
plt.yscale('log')
plt.xlabel(r'leading b quark $p_T$')
plt.suptitle("EWKWMinus2Jets")
plt.legend()
```

```python
had_bquarks = ak.firsts(had_events.FatJet.pt)
lep_bquarks = ak.firsts(lep_events.FatJet.pt)
```

```python
plt.hist(had_bquarks[had_bquarks>0],histtype='step',bins=np.linspace(150,5000,50),label='WToQQ',density=True)
plt.hist(lep_bquarks[lep_bquarks>0],histtype='step',bins=np.linspace(150,5000,50),label='WToLNu',density=True)
plt.yscale('log')
plt.xlabel(r'Fat jet $p_T$')
plt.suptitle("EWKWMinus2Jets")
plt.legend()
```
