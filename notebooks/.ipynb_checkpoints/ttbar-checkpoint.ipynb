```python
import os, sys
import subprocess
import json
import uproot3
import awkward as ak
import numpy as np
from coffea import processor, util, hist
import matplotlib.pyplot as plt
import pickle
```

```python
year='2016APV'

# Check if pickle exists     
picklename = '../'+year+'/ttbar-templates.pkl'
if not os.path.isfile(picklename):
    print("You need to create the pickle")
```

```python
# Read the histogram from the pickle file
templates = pickle.load(open(picklename,'rb')).sum('genflavor','pt1').integrate('region','signal-vbf').integrate('systematic','nominal')
```

```python
mjjbin1 = templates.integrate('mjj',int_range=slice(1000,2000))
mjjbin2 = templates.integrate('mjj',int_range=slice(2000,13000))
```

```python
fig = plt.subplots(1,1,figsize=(5,5))
hist.plot1d(mjjbin1.integrate('ddb1',int_range=slice(0,0.64)),overlay='process')
plt.title(r'$m_{jj}$ bin 1, DDB fail, '+ year)
plt.savefig('../ttbar-plots/'+year+'failmjjbin1')
plt.show()

fig = plt.subplots(1,1,figsize=(5,5))
hist.plot1d(mjjbin1.integrate('ddb1',int_range=slice(0.64,1)),overlay='process')
plt.title(r'$m_{jj}$ bin 1, DDB pass, '+ year)
plt.savefig('../ttbar-plots/'+year+'passmjjbin1')
plt.show()
```

```python
fig = plt.subplots(1,1,figsize=(5,5))
hist.plot1d(mjjbin2.integrate('ddb1',int_range=slice(0,0.64)),overlay='process')
plt.title(r'$m_{jj}$ bin 2, DDB fail, '+ year)
plt.savefig('../ttbar-plots/'+year+'failmjjbin2')
plt.show()

fig = plt.subplots(1,1,figsize=(5,5))
hist.plot1d(mjjbin2.integrate('ddb1',int_range=slice(0.64,1)),overlay='process')
plt.title(r'$m_{jj}$ bin 2, DDB pass, '+ year)
plt.savefig('../ttbar-plots/'+year+'passmjjbin2')
plt.show()
```
