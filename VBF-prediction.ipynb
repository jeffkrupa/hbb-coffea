```python
import os, sys
import subprocess
import json
import uproot
import awkward as ak

from coffea import processor, util, hist
from coffea.nanoevents import NanoEventsFactory, NanoAODSchema
from boostedhiggs import VBFTruthProcessor

from datetime import datetime
```

```python
for year in ['2017']:
    index='VBFHToBB'
    this_file = 'infiles/'+str(year)+'_'+index+'.json'
    outfile = 'outfiles-higgs/'+str(year)+'_dask_'+index+'.coffea'

    p = VBFTruthProcessor(year=year,jet_arbitration='ddb')
    args = {'savemetrics':True, 'schema':NanoAODSchema}
    
    output = processor.run_uproot_job(
        this_file,
        treename="Events",
        processor_instance=p,
        executor=processor.dask_executor,
        executor_args={"skipbadfiles": 1,
                       "schema": processor.NanoAODSchema,
                       "treereduction": 2},
        chunksize=100000)

    util.save(output, outfile)
    print("saved " + outfile)
```
